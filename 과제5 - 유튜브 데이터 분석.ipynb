#### 1-2. 조회수 상위 100개 동영상에서 많이 사용된 태그 찾기

```python
def string_handle(x): #문자열 처리 함수
    x = x.lower() #소문자로 변경
    x = x.replace('"','') #"제거
    x = x.replace('#','') ##제거
    x = x.replace('_', ' ') #_를 공백으로 변경
    x = x.strip() #양쪽 공백 제거
    return x
```

```python
tag100 = []
for tags in kr100['tags']:
    tag_lst = tags.split('|')
    tag100.extend(tag_lst)
```

```python
handled_tag100 = [string_handle(x) for x in tag100]
```

```python
tag100_count=pd.DataFrame(data={'tag':handled_tag100}).groupby('tag').size().reset_index().rename(columns={0:'count'}).sort_values('count',ascending=False)
tag100_count.head(15)
```

많이 사용된 태그 상위 15개에서 pop을 제외한 14개 모두 marvel이나 방탄소년단과 연관된 태그이다. 17~18년도 한국 유튜브 트랜드는 마블 영화인 어밴져스와 방탄소년단이 지배적이였던 것으로 보인다.

#### 1-3. 캐나다, 미국, 영국의 조회수 상위 100개 영상의 태그 트렌드 분석

```python
ca=pd.read_csv('CAvideos.csv')
us=pd.read_csv('USvideos.csv')
gb=pd.read_csv('GBvideos.csv')
```

```python
def top_tag10(data):
    df100 = data.sort_values('views', ascending=False).iloc[:100,:]
    tag100 = []
    for tags in df100['tags']:
        tag_lst = tags.split('|')
        tag100.extend(tag_lst)
    handled_tag100 = [string_handle(x) for x in tag100]
    tag100_count=pd.DataFrame(data={'tag':handled_tag100}).groupby('tag').size().reset_index().rename(columns={0:'count'}).sort_values('count',ascending=False)
    return tag100_count.iloc[:10,:]
```

```python
top_tag10(ca)
```

캐나다도 한국과 마찬가지로 marvel의 어밴저스가 지배적인것을 볼 수 있다. 더불어 Ed Sheeran의 노래도 인기를 끈 것으로 보인다.

```python
top_tag10(us)
```

### 3. like 와 dislike (호감/비호감) 비율 구한 뒤, 어떤 카테고리인지 파악하기

```python
# 패키지 불러오기
from pandas import *
import pandas as pd
```

```python
# 현재 위치 확인 
import os 
os.getcwd() # 현재 디렉토리 위치 확인 
```

```python
# 데이터 불러오기 
input_fd = open('KRvideos.csv',errors = 'backslashreplace',encoding='utf-8')
KRvideos = pd.read_csv(input_fd)

KRvideos.head()
```

```python
# 카데고리 설명 json 파일 불러오기 
import json
from pprint import pprint

with open('KR_category_id.json') as data_file:    
    data = json.load(data_file)
    
pprint(data)
```

```python
# 카테고리 설명 데이터프레임 만들기 
ID = []
title = []

for i in range(len(data["items"])):
    ID.append(data["items"][i]["id"])
    title.append(data["items"][i]["snippet"]["title"])

category_desc = pd.DataFrame({'category_id':ID, 
                             'category': title})
category_desc
```

31가지 카테고리가 있다.

```python
# 문자열 변환 
category_desc["category_id"] = category_desc["category_id"].apply(str)
KRvideos["category_id"] = KRvideos["category_id"].apply(str)

# 기존 KR 데이터프레임에 카테고리 설명 데이터프레임을 합친 뒤, KR에 저장하기 
KR = pd.merge(KRvideos, category_desc, on="category_id", how='left')
```

### [좋아요(likes) 상위 25%의 카테고리 분포]

```python
# 좋아요 내림차순으로 정렬 후 KR_likes에 저장 
KR_likes = KR.sort_values(by=['likes'], ascending=False)

# 좋아요 상위 25% 추출 후 KR_25에 저장 
KR_likes25 = KR_likes.iloc[1:round(len(KR_likes)*0.25)]
KR_likes25.head(3)
```

```python
# 좋아요 상위 25%의 카테고리 분포 살펴보기 
KR_likes25['category'].astype(object)
KR_likes25.category.value_counts()
```

#### 좋아요 수가 많은(상위 25% 수준) 유투브 동영상의 카데고리 분포를 살펴본 결과, Entertainment가 2377개로 가장 많았고, Music과 People&Blogs가 각각 1153개와 1130개로 그 뒤를 이었으며, Comedy도 1031개로 많았다. 

```python
round(KR_likes25.category.value_counts()/len(KR_likes25)*100, 2)
```

#### Entertainment 27.51%, Music 13.34%, People & Blogs 13.08%, Comedy 11.93%, News & Politics 11.29%로 집계되었다.

#### 즉, 한국인에게 좋아요를 많이 받은 유투브 영상은 주로 연예, 음악, v-log 콘텐츠인 것으로 드러났으며, 교육, 과학기술, 여행, 자동차 관련 콘텐츠는 좋아요 수가 많지 않았다.

### [싫어요(dislikes) 상위 25%의 카테고리 분포]

```python
# 같은 방식으로 싫어요 상위 25%의 카테고리 분포를 살펴보자

KR_dislikes = KR.sort_values(by=['dislikes'], axis=0, ascending=False)
KR_dislikes25 = KR_dislikes.iloc[1:round(len(KR_dislikes)*0.25)]
KR_dislikes25['category'].astype(object)
KR_dislikes25.category.value_counts()
```

#### 싫어요 수가 많은 카테고리는 Entertainment(2548개), News & Politics(1811개), People & Blogs(1327개), Music(759개) 순인 것으로 나타났다. 


```python
round(KR_dislikes25.category.value_counts()/len(KR_dislikes25)*100, 2)
```

#### 비율을 살펴보면, Entertainment가 29.49%로 가장 많았고, News & Politics(20.96%)과 People & Blogs(15.36%)로 그 뒤를 이었으며, Howto & Style, Education, Pets & Animals, Science & Technology, Travel & Events, Autos & Vehicles, Shows는 싫어요를 많이 받지 않았다. 

```python
KR_likes25["title"].head(20)
```

```python
KR_dislikes25["title"].head(20)
```

```python
total_25 = pd.concat([KR_likes25, KR_dislikes25])
total_25["duplicated"] = total_25.duplicated()
```

```python
# 좋아요 상위 25% 영상 개수
print(KR_likes25.shape[0])
# 싫어요 상위 25% 영상 개수
print(KR_dislikes25.shape[0])
# 겹치는 영상 개수 
print(len(total_25[total_25["duplicated"]== True]))
```

```python
8641-5943
```

#### 상위 25%에 해당하는 영상의 개수는 각각 8641개이며, 그 중 5943개의 영상이 겹치고 중복되지 않는 영상의 개수는 각각 2698개이다. 따라서 겹치는 영상이 많기 때문에 좋아요 카테고리와 싫어요 카테고리의 순위가 일치한다고 볼 수 있다. 
#### 좋아요 최상위권과 싫어요 최상위권은 다소 중복되는 경향이 있었는데, 제목을 살펴보면 대부분 BTS(방탄소년단)의 공식 뮤직비디오였다. 이러한 현상은 BTS 유행에 따른 호기심과 대중의 적극적 평가 등이 영향을 미쳤을 것으로 예상된다. 

#### <결론>

#### 한국 Youtube 동영상 중에서 가장 좋아요 혹은 싫어요가 많은 카테고리는 엔터테인먼트였으며, 좋아요가 많은 카테고리 순위와 싫어요가 많은 카테고리 순위가 유사했다. 이는 좋아요 순위가 높은 영상과 싫어요 순위가 높은 영상이 겹치기 때문인 것으로 보인다. Entertainment와 People&Blogs 콘텐츠에 대해 반응이 활발하게 일어나며, 뉴스나 정치적 사안을 다룬 영상에 대해서는 싫어요 수가 많다는 것을 알 수 있었다.
